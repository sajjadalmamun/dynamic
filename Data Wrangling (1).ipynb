```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv', header = None)
```

```python
df.head()
```

```python
header = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df.columns = header
```

```python
df.head()
```

```python
df.replace('?', np.nan, inplace = True)
```

```python
df.head()
```

```python
df.isnull()
```

```python
for column in df.isnull().columns.values.tolist():
    print(column)
    print(df.isnull()[column].value_counts())
    print("")
```

```python
MeanNormalizedLosses = df['normalized-losses'].astype('float').mean(axis=0)
df.replace(np.nan, MeanNormalizedLosses, inplace = True)
```

```python
df.head()
```

```python
MeanBore = df['bore'].astype('float').mean(axis=0)
df.replace(np.nan, MeanBore, inplace = True)
```

```python
MeanStroke = df['stroke'].astype('float').mean(axis=0)
df.replace(np.nan, MeanStroke, inplace = True)
```

```python
MeanHorsepower = df['horsepower'].astype('float').mean(axis=0)
df.replace(np.nan, MeanHorsepower, inplace = True)
```

```python
MeanPeakRpm = df['peak-rpm'].astype('float').mean(axis=0)
df.replace(np.nan, MeanPeakRpm, inplace = True)
```

```python
df.replace(np.nan, np.nan, inplace = True)
```

```python
df.dropna(subset = ['price'], axis = 0, inplace = True)
```

```python
df.reset_index(drop = True, inplace = True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax
```

```python
df['num-of-doors'].replace(122.0, 'four', inplace = True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df.dtypes
```

```python
df['price'].astype('float')
```

```python
df[['peak-rpm', 'bore', 'stroke', 'price']] = df[['peak-rpm', 'bore', 'stroke', 'price']].astype(float)
```

```python
df[['normalized-losses']] = df[['normalized-losses']].astype('int')
```

```python
df.dtypes
```

```python
df['city-mpg'] = 235/ df['city-mpg']
```

```python
df.rename(columns= {'highway-mpg':'highway-L/100km'}, inplace = True)
```

```python
df.head()
```

```python
df['length']= df['length']/df['length'].max()
```

```python
df['width']= df['width']/ df['width'].max()
```

```python
df['height'] = df['height']/df['height'].max()
```

```python
df[['horsepower']]= df[['horsepower']].astype('int')
```

```python
df['horsepower'].replace(122.0,122, inplace = True)
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins = np.linspace(min(df['horsepower']), max(df['horsepower']), 4)
group = ['low', 'medium', 'high']
df['HorsepowerBinned'] = pd.cut(df['horsepower'],bins, labels = group, include_lowest = True)
```
